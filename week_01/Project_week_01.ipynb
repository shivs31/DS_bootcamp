```python
continents = pd.read_csv('../data/continents.csv', index_col=0)
continents
```

```python
continents.index
```

```python
continents.index.name = 'country'
```

```python
continents = continents.reset_index()
continents
```

```python
continents[['continent', 'country']] = continents["country"].apply(lambda x: pd.Series(str(x).split(";")))
continents
```

```python
df = fert.merge(population)
df = df.merge(life)
df = df.merge(continents)
df
```

```python
import seaborn as sns

df_subset = df.loc[df['year'] == 2015]
sns.scatterplot(
    x= 'life_expectancy',
    y ='fertility_rate',
    data=df_subset,
    hue='continent',
    alpha=1,
    size='population'
)
plt.axis([0, 90, 0, 8]);
```

```python
import imageio
images = []
for i in range(1960,2016): 
    df_subset1 = df.loc[df['year'] == i]
    p=sns.scatterplot(
       x= 'life_expectancy',
       y ='fertility_rate',
       data=df_subset1,
       hue='continent',
       alpha=1,
       size='population',
       sizes=(2,2000)
    )
    h,l = p.get_legend_handles_labels()
    plt.legend(h[0:7], l[0:7], loc='upper left', borderaxespad=0)
    plt.axis([0.0, 85, 0, 8])
    plt.savefig('lifexp_{}.png'.format(i))
    plt.close()   
```

```python
import imageio

images = []

for i in range(1960,2016):
    shivani = 'lifexp_{}.png'.format(i)
    images.append(imageio.imread(shivani))

imageio.mimsave('output.gif', images, fps=15)
```
