```python
import pandas as pd
```

Loading the file

```python
df = pd.read_csv('../data/gapminder_total_fertility.csv', index_col=0)
df
```

Selecting 3 years from the columns of DataFrame

```python
df = df[['1950', '1955', '2000']]
df
```

Select 3 countries from the DataFrame index

```python
df = df.loc[['Germany', 'India', 'Bulgaria', 'Kenya']]
df
```

Draw a bar plot using the years as bar groups:

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(5,7))
df.plot.bar()
plt.savefig('bars.png');
```
